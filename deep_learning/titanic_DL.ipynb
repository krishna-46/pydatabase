```python
import tensorflow as tf
import pandas as pd
from tensorflow.keras.layers import Dense,Dropout,Flatten
```

```python
df = pd.read_csv('Dataset/titanic.csv',usecols=['PassengerId' , 'Survived' , 'Pclass' , 'Sex' , 'Age' , 'SibSp' , 'Parch' , 'Embarked'])
df.head()
```

```python
df.isnull().sum()
```

```python
df.fillna(method='ffill',inplace=True)
```

```python
df.isnull().sum()
```

```python
x = df.iloc[:,[0,2,4,5,6]]
x.shape
```

```python
y = df['Survived']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train , x_test , y_train , y_test=train_test_split(x,y,train_size=0.8,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer

from sklearn.preprocessing import OneHotEncoder , OrdinalEncoder
```

```python
trf = ColumnTransformer(transformers=[
    ('tr1',OneHotEncoder(sparse=False) , ['Sex' ,'Embarked'])
    
],remainder='passthrough')
```

```python
pd.get_dummies(df,columns=['Sex' , 'Embarked'])
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dtc = DecisionTreeClassifier()
```

```python
dtc.fit(x_train ,  y_train)
```

```python
y_pred = dtc.predict(x_test)
y_pred
```

```python
df.drop_duplicates(inplace=True)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_pred,y_test)
```

```python
model = tf.keras.models.Sequential([
    Dense(units=5,use_bias=True),
    Dense(128 , activation='relu',use_bias=True),
    Dense(10,activation='softmax',use_bias=True)

])
```

```python
model.compile(optimizer='adam',loss='sparse_categorical_crossentropy',metrics=['accuracy'])
```

```python
model.fit(x_train , y_train , epochs=100000)
```

```python
model.evaluate(x_test,y_test)
```
